## matplotlib workflow example:

```python
#0. import matplotlib and setup ready for jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt
# 1.prepare data.
x = [1,2,3,4]
y = [11,22,33,44]
# 2. setup plot
fig,ax=plt.subplots(figsize=(10,10))# width, height
#3.plot data
ax.plot(x,y)
#4.customize plot
ax.set(title='simple plot',
       xlabel='x-axis',
       ylabel='y-axis')
#5. show and save ( save the whole figure)
plt.show()
fig.savefig('img/sample-plot.png')# you can right click and save image as
```

```python
#make a plot from a dictionary
nut_butter_prices = { 'Almond butter':10 ,
                     'Peanut butter': 8 ,
                     'Cashew butter':12}
fig,ax=plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values()) # you can have height=, but not x=. ie the order of args kwargs etc.
ax.set(title='Basil Nut store',
       xlabel='TYPES',
       ylabel='PRICE ($)')
plt.show()
```

```python
#Make some data, then use it to create histogram.
x = np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x);
plt.show()
```

## Two options for subplots

```python
# subplot option 1   ---Recommended.
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                    ncols=2,
                                    figsize=(10,5))
#plot to each different axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
plt.show()
```

```python
#subplots option 2
fig, ax=plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10,5))
#plot to each different by indexing.
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000))
plt.show()
```

# which one to use? (pyplot vs object oriented API method)
## when plotting sth quickly, use the pyplot.
## when plotting sth more advanced, use the oo method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
len(over_50)
```

```python
#pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target')
plt.show()
```

```python
# oo method mixed with pyplot method
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
            ax=ax)
#ax.set_xlim([45,100])- adjust our x-axis to view from 45 to 100ax.set_ylim([50,400])
#ax.set_ylim([50,600])- adjusts y as instructed.
plt.show()
```

```python
# oo method from scratch
fig, ax= plt.subplots(figsize=(10,6))
# plot the data
scatter= ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])
#customize the plot
ax.set(title='Heart Disease and Cholesterol levels',
       xlabel='Age',
       ylabel='cholesterol')
#add a legend
ax.legend(*scatter.legend_elements(), title='Target')
#add a horizontal line
ax.axhline(over_50['chol'].mean(),linestyle='--')
plt.show()
```

```python
over_50.head()
```

```python
#subplots in OO method
#subplot of age,chol,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10))
##add data to ax0
scatter= ax0.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])
#customize the plot
ax0.set(title='Heart Disease and Cholesterol levels',
       xlabel='Age',
       ylabel='cholesterol')
#add a legend
ax0.legend(*scatter.legend_elements(), title='Target')
#add a horizontal line
ax0.axhline(over_50['chol'].mean(),linestyle='--')

##add data to ax1
scatter1=ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'])
#customize the plot
ax1.set(title='Heart Disease and Thalach',
        xlabel='Age',
        ylabel='Max Heart Rate')
#add legend
ax1.legend(*scatter1.legend_elements(),title='Target')
#add a horizontal line- here is mean
ax1.axhline(over_50['thalach'].mean(),linestyle='--')
plt.show()
```

```python
#subplots in OO method
#subplot of age,chol,thalach---SHAREX=True, to ensure they share the xlabel age.
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10),
                          sharex=True)
##add data to ax0
scatter= ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                    c=over_50['target'])
#customize the plot
ax0.set(title='Heart Disease and Cholesterol levels',
        ylabel='cholesterol')
#add a legend
ax0.legend(*scatter.legend_elements(), title='Target')
#add a horizontal line
ax0.axhline(over_50['chol'].mean(),linestyle='--')

##add data to ax1
scatter1=ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'])
#customize the plot
ax1.set(title='Heart Disease and Thalach',
         xlabel='Age',
        ylabel='Max Heart Rate')
#add legend
ax1.legend(*scatter1.legend_elements(),title='Target')
#add a horizontal line- here is mean
ax1.axhline(over_50['thalach'].mean(),linestyle='--')
#add a title to the entire figure.
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold')
plt.show()
```

```python
#set the style
plt.style.use('seaborn-v0_8-whitegrid')
# oo method from scratch
fig, ax= plt.subplots(figsize=(10,6))
# plot the data
scatter= ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'],
                    cmap='winter')#changes the color scheme ie modifies. Others include summer, plasma etc. Prefer winter
#customize the plot
ax.set(title='Heart Disease and Cholesterol levels',
       xlabel='Age',
       ylabel='cholesterol')
#add a legend
ax.legend(*scatter.legend_elements(), title='Target')
#add a horizontal line
ax.axhline(over_50['chol'].mean(),linestyle='--')
plt.show()
```

```python
#customizing the X and Y axis limitations.
#subplots in OO method
#subplot of age,chol,thalach---SHAREX=True, to ensure they share the xlabel age.
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10),
                          sharex=True)
##add data to ax0
scatter= ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                    c=over_50['target'],
                    cmap='winter')
#customize the plot
ax0.set(title='Heart Disease and Cholesterol levels',
        ylabel='cholesterol')
#change the x axis limits
ax0.set_xlim([50,80])
#add a legend
ax0.legend(*scatter.legend_elements(), title='Target')
#add a horizontal line
ax0.axhline(over_50['chol'].mean(),linestyle='--')

##add data to ax1
scatter1=ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'],
                    cmap='winter')
#customize the plot
ax1.set(title='Heart Disease and Thalach',
         xlabel='Age',
        ylabel='Max Heart Rate')
#change the x & y axes limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
#add legend
ax1.legend(*scatter1.legend_elements(),title='Target')
#add a horizontal line- here is mean
ax1.axhline(over_50['thalach'].mean(),linestyle='--')
#add a title to the entire figure.
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold')
plt.show()
```
